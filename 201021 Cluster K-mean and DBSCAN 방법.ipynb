```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
X, y_true = make_blobs(n_samples=400, centers=4, cluster_std=0.60, random_state=0)
```

```python
plt.scatter(X[:,0],X[:,1], s=30)
```

```python
y_true
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
```

```python
y_kmeans = kmeans.predict(X)
```

```python
y_kmeans
```

```python
list(y_kmeans).count(0)
```

```python
list(y_true).count(0)
```

```python
centers = kmeans.cluster_centers_
plt.scatter(X[:,0],X[:,1], c=y_kmeans, s=20, cmap='viridis')
plt.scatter(centers[:,0], centers[:,1], c='black', s=100, alpha=0.6)
```

```python
sns.scatterplot(X[:,0], X[:,1], hue=y_kmeans)
```

```python
from sklearn.datasets.samples_generator import make_moons
X, y = make_moons(200, noise = 0.05, random_state=0)
plt.scatter(X[:,0],X[:,1], s=30)
```

```python
list(y).count(0)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
y_kmeans = kmeans.predict(X)
```

```python
centers = kmeans.cluster_centers_
plt.scatter(X[:,0],X[:,1], c=y_kmeans, s=20, cmap='viridis')
plt.scatter(centers[:,0], centers[:,1], c='black', s=100, alpha=0.6)
```

```python
#   
sns.scatterplot(X[:,0], X[:,1], hue=y)
```

```python
sns.scatterplot(X[:,0], X[:,1], hue=y_kmeans)
```

```python
from sklearn.cluster import DBSCAN
#  ,    
DBS = DBSCAN(eps = 0.3, min_samples=15)
```

```python
y_prediction = DBS.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1], c=y_prediction, s=20, cmap='viridis')
```
