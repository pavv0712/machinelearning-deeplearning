```python
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions
import numpy as np
import os
import glob
```

```python
model = VGG16()
```

```python
img_path='C:/Users/ASIA_01/Desktop/이미지'
```

```python
glob.glob(os.path.join(img_path, '*.*'))
```

```python
filenames = glob.glob(os.path.join(img_path, '*.*'))
```

```python
filename = filenames[10]
img = image.load_img(filename, target_size=(224,224))
X = image.img_to_array(img)
```

```python
X=np.expand_dims(X, axis=0)
```

```python
features = model.predict(X)
```

```python
label = decode_predictions(features)
```

```python
label[0][0][1], label[0][0][2]
```

```python
for filename in glob.glob(os.path.join(img_path, '*.*')):
    img = image.load_img(filename, target_size=(224,224))
    X = image.img_to_array(img)
    X=np.expand_dims(X, axis=0)
    features = model.predict(X)
    label = decode_predictions(features)
    print('{}={},{}%'.format(filename,label[0][0][1], label[0][0][2]))
```
