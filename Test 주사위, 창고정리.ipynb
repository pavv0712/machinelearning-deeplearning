```python
f = open('C:/Users/ASIA_01/Downloads/주사위게임1.txt', 'r')
n = int(f.readline())
high_prize = 0
for i in range(n):
    temp = f.readline().split()
    temp.sort()
    a,b,c = map(int, temp)
    if a==b and b==c: #주사위눈 3개
        prize = 1000+(a*1000)
    elif a==b or b==c: #주사위눈 2개 
        prize = 1000+(b*100)
    else: #주사위눈 모두 다름
        prize = c*100
    
    if prize > high_prize:
        high_prize = prize
print(high_prize)
```
