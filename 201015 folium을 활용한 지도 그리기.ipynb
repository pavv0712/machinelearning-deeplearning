```python
import folium
```

```python
m = folium.Map(location= [37.562531, 126.992568], zoom_start=17)
```

```python
m
```

```python
#marker 출력
m = folium.Map(location= [37.562531, 126.992568], zoom_start=17, tiles ='OpenStreetMap')
tooltip = 'please Click'
folium.Marker(location= [37.566743, 126.992593], popup='<i>up<i>', tooltip=tooltip).add_to(m)
folium.Marker(location= [37.562531, 126.992568], popup='<i>center<i>', tooltip=tooltip).add_to(m)
folium.Marker(location= [37.561519, 126.994037], popup='<i>down<i>', tooltip=tooltip).add_to(m)


m
```

```python
#icon 출력하기
m = folium.Map(location= [37.562531, 126.992568], zoom_start=17, tiles ='OpenStreetMap')
tooltip = 'please Click'
folium.Marker(location= [37.562531, 126.992568], popup='<i>center<i>', icon=folium.Icon(color='green', icon='star')).add_to(m)
```

```python
m
```

```python
#Circle
m = folium.Map(location= [37.562531, 126.992568], zoom_start=17, tiles ='OpenStreetMap')
folium.CircleMarker([37.562531, 126.992568], radius=150, color='#74F183', fill=True, fill_color='red').add_to(m)
m
```
